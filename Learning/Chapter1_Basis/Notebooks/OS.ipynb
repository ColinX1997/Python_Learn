ZiP - 4G

```python
import zipfile
import os

source_folder = 'C:\\MES-Component\\_workspace\\python\\Learning\\temp'
target_folder = 'C:\\MES-Component\\_workspace\\python\\Learning\\resource\\'

try:
    zipfile = zipfile.ZipFile(target_folder + 'ziptest.zip','w',zipfile.ZIP_DEFLATED)
    for file in os.listdir(source_folder):
        file_path= os.path.join(source_folder, file)
        print(file_path,os.path.isfile(file_path))
        if os.path.isfile(file_path):
            zipfile.write(file_path)
finally:
    zipfile.close()
```
